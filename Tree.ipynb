```python
from anytree import Node, RenderTree

def add_nodes(nodes, parent, child):
  if parent not in nodes:
    nodes[parent] = Node(parent)
  if child not in nodes:
    nodes[child] = Node(child)
  nodes[child].parent = nodes[parent]
```

```python
nodes = {}
for parent, child in zip(df['아비'],df['kpn']):
  add_nodes(nodes, parent, child)

print(nodes)

roots = list(df[~df['아비'].isin(df['kpn'])]['아비'].unique())

print(roots)

for root in roots[0:]:
  for pre, _, node in RenderTree(nodes[root]):
    print("%s%s" % (pre, node.name))
```

```python
parent_index = []
for i in df['아비']:
  if df.index[(df['kpn']) == i].tolist():
    index_num = df.index[(df['kpn']) == i].tolist()[0]
  else:
    index_num = 0
  parent_index.append(index_num)

df['parent_index'] = parent_index
df_data = df.loc[:,['kpn','parent_index']]
df_data.index = df_data.index+1
df_data['kpn_index'] = df_data.index
df_data = df_data.fillna('hanwoo')
df_data = df_data[['kpn_index','kpn','parent_index']]
df_data['kpn_index'] = (df_data['kpn_index']).apply(str)
df_data['parent_index'] = df_data['parent_index'].apply(str)

new_row = pd.DataFrame([['0','hanwoo','hi']],columns=df_data.columns)
df_data = pd.concat([df_data.iloc[:0], new_row, df_data.iloc[0:]],ignore_index = True)

df_data
```

```python
hanwoo = []
for i in df_data.values:
  hanwoo.append(tuple(i))
print(hanwoo)
```

```python
import pprint
from collections import defaultdict
import json

def show_val(title, val):
    sep = '-' * len(title)
    print("\n{0}\n{1}\n{2}\n".format(sep, title, sep))
    pprint.pprint(val) 

parents = defaultdict(list)
for p in hanwoo:
    parents[p[2]].append(p)

show_val("parents", parents)

def buildtree(t=parents, parent_eid=''):
    """
    Given a parents lookup structure, construct
    a data hierarchy.
    """
    parent = parents.get(parent_eid, None)
    if parent is None:
        return t
    for eid, name, mid in parent:
        report = { 'name': name }
        if t is None:
            t = report
        else:
            reports = t.setdefault('reports', [])
            reports.append(report)
        buildtree(report, eid)
    return t

data = buildtree(parents)
show_val("data", data)

show_val("JSON", json.dumps(data))
```
